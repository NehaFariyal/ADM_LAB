```python
import pandas as pd
```

```python
df = pd.read_csv('/content/customer_sales_data.csv')
print("Dataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display basic statistics
print("Summary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(df['Sales_Amount'], bins=10, kde=True, color='red')
plt.title("Distribution of Sales Amounts")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# Option 1: Sales_ID vs Sales_Amount
# This shows how Sales Amount varies across different Sales IDs.
sns.scatterplot(x=df['Sales_ID'], y=df['Sales_Amount'], hue=df['Product_Category'])

# Option 2: Quantity_Sold vs Sales_Amount
# This shows the relationship between the quantity of products sold and the sales amount.
sns.scatterplot(x=df['Quantity_Sold'], y=df['Sales_Amount'], hue=df['Product_Category'])

# Option 3: Customer_Age vs Sales_Amount
# This shows how Sales Amount is influenced by Customer Age.
sns.scatterplot(x=df['Customer_Age'], y=df['Sales_Amount'], hue=df['Product_Category'])
```

```python
numeric_df = df.select_dtypes(include=['number']) # Selects only numeric columns
```

```python
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
